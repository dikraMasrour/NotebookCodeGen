# First strategy 

```python
def construct_pairs(df,i,new_df,limit):

    if df.loc[i]['cell_type'] == 'markdown' and i < limit:
        grp_md(i,df,new_df,limit)

    elif df.loc[i]['cell_type'] == 'code' and i < limit:
        grp_cd(i,df,new_df,limit)       
   
```

```python
def grp_md(i,df,new_df,limit):
    
    cell_md = ""
    
    while i < limit and df.loc[i]['cell_type'] == 'markdown':
        print(i)
        cell_md += str(df.loc[i]['source']) +" "
        i += 1

    new_df.loc[len(new_df)] = ['markdown',cell_md,df.loc[i]['title']]


    if i < limit :
        construct_pairs(df,i,new_df,limit)
```

```python
def grp_cd(i,df,new_df,limit):
    cell_cd = ""

    
    while i < limit and df.loc[i]['cell_type'] == 'code':
        print(i)
        cell_cd += str(df.loc[i]['source']) +" "
        i += 1

    new_df.loc[len(new_df)] = ['code',cell_cd,df.loc[i]['title']]



    if i < limit :
        construct_pairs(df,i,new_df,limit)
```

```python
new_df = pd.DataFrame(columns=['cell_type','source','title'])
```

```python
List_ntb = []
for i in df.index:
    List_ntb.append(df.loc[i]['title'])

List_ntb_only = np.unique(List_ntb)
```

```python
for ntb in List_ntb_only:
    construct_pairs(df[df['title'] == ntb].reset_index(drop=True),0,new_df,len(df[df['title'] == ntb]))
```
