```python
train=[io.imread('train/'+str(i)+'.Bmp',as_gray=True) for i in range(1, 6284)]
plt.imshow(train[0])
```

```python
final_train=[cv2.resize(image,(28,28)).flatten() for image in train]
```

```python
labels=pd.read_csv('/kaggle/input/street-view-getting-started-with-julia/trainLabels.csv')
labels.head()
```

```python
finaly=np.array(final_train)
```

```python
finaly.shape
```

```python
trainy= labels['Class'].iloc[:]
```

```python
trainx = finaly
```

```python
model=KNeighborsClassifier()
model.fit(trainx,trainy)
```

```python
import zipfile

Dataset = "test"

# Will unzip the files so that you can see them..
with zipfile.ZipFile("/kaggle/input/street-view-getting-started-with-julia/"+Dataset+".zip","r") as z:
    z.extractall(".")

import re    

def sorted_alphanumeric(data):
    convert = lambda text: int(text) if text.isdigit() else text.lower()
    alphanum_key = lambda key: [ convert(c) for c in re.split('([0-9]+)', key) ] 
    return sorted(data, key=alphanum_key)    

lst = sorted_alphanumeric(os.listdir('test/'))


```

```python
lst
```

```python
lab=[]
test=[]
import os
for i in lst:
    test.append(io.imread('test/'+i,as_gray=True))
    lab.append(i.split('.')[0])
```

```python
plt.imshow(test[0])
```

```python
#test_img=np.array([cv2.resize(image,(28,28)) for image in test])
#test_img=test_img[:,:,:,np.newaxis]
#test_img.shape

final_test=[cv2.resize(image,(28,28)).flatten() for image in test]
f_test = np.array(final_test)
f_test.shape
```

```python
res = model.predict(f_test)
```

```python
res
```

```python
sub=pd.DataFrame({'ID':lab,'Class':res})
```
